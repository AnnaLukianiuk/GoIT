```python
import pandas as pd
```

```python
df = pd.read_csv("applications.csv")
display(df.head())
```

```python
print(df.shape)
```

```python
df.info()
```

```python
print(df.columns)
```

```python
# 1.1 Прибери дублікати applicant_id

df.drop_duplicates(subset='applicant_id', inplace=True)
df.info()
```

```python
# 1.2 В полі 'External Rating' заповни відсутні значення нулями

display(df[df['External Rating'].isna()])

df['External Rating'].fillna(0, inplace=True)
```

```python
# 1.3 В полі 'Education level' заповни відсутні значення текстом “Середня”

df['Education level'] = df['Education level'].fillna('Середня')
df.info()
```

```python
# 2. Додай до цього DataFrame дані з файлу industries.csv, а саме, рейтинги індустрій

df_ind = pd.read_csv("industries.csv")
display(df_ind)
```

```python
merged = pd.merge(left=df,
                  right=df_ind,
                  on='Industry',
                  how='left')

display(merged.head())
```

```python
merged.info()
```

```python
# 3. Розрахуй рейтинг заявки за наступними умовами:

# Якщо вік між 35 та 55, то до рейтингу додається 20 балів

filter_1 = ((merged['Age'] >= 35) & (merged['Age'] <= 55)) * 20
print(filter_1)
```

```python
# Якщо заявка була подана не у вихідні, то до рейтингу додається 20 балів

merged['Applied at'] = pd.to_datetime(merged['Applied at'], dayfirst = True, format='mixed')
print(merged['Applied at'])
```

```python
filter_2 = ((merged['Applied at'].dt.day_name() != "Saturday") & (merged['Applied at'].dt.day_name() != "Sunday")) * 20
print(filter_2)
```

```python
# Якщо одружений, то до рейтингу додається 20 балів

filter_3 = (merged['Marital status'] == "Married") * 20
print(filter_3)
```

```python
# Якщо знаходиться в Києві чи області, то до рейтингу додається 10 балів

filter_4 = (merged['Location'] == "Київ чи область") * 10
print(filter_4)
```

```python
# Якщо 'External Rating' більше чи дорівнює 7, то до рейтингу додається 20 балів

filter_5 = (merged['External Rating'] >= 7) * 20
print(filter_5)
```

```python
# Якщо 'External Rating' менше чи дорівнює 2, то з рейтингу віднімається 20 балів

filter_6 = (merged['External Rating'] <= 2) * (-20)
print(filter_6)
```

```python
# Рейтинг - сума оцінок заявки по 6 критеріям

merged['Rating'] = merged['Score'] + filter_1 + filter_2 + filter_3 + filter_4 + filter_5 + filter_6
print(merged['Rating'])
```

```python
# Рейтинг дорівнює нулю, якщо відсутнє значення 'Amount' або якщо 'External Rating' дорівнює нулю

merged.loc[merged['Amount'].isna() | (merged['External Rating'] == 0), 'Rating'] = 0
merged[(merged['Amount'].isna()) | (merged['External Rating'] == 0)][['Amount', 'External Rating', 'Rating']]
```

```python
# Рейтинг має бути числом від 0 до 100

print(merged['Rating'].min())

merged.loc[merged['Rating'] < 0, 'Rating'] = 0
print(merged['Rating'].min())
```

```python
print(merged['Rating'].max())

merged.loc[merged['Rating'] > 100, 'Rating'] = 100
print(merged['Rating'].max())
```

```python
# 4. В результуючій таблиці залиши лише заявки з рейтингом більше нуля, ці заявки вважатимуться прийнятими

result_table = merged[merged['Rating'] > 0]

display(result_table.sort_values(by='Rating').head())
```

```python
# 5. Дані з результуючої таблиці згрупуй за тижнем подачі заявки та виведи на графік середній рейтинг прийнятих заявок

result_table.loc[:, 'Applied at'] = result_table['Applied at'].dt.isocalendar().week
display(result_table.head())
```

```python
weekly_avg_rating = result_table.groupby('Applied at')[['Rating']].mean().round(1)
display(weekly_avg_rating)
```

```python
print(weekly_avg_rating.plot(kind='bar', title='Weekly average rating of accepted applications'))
```

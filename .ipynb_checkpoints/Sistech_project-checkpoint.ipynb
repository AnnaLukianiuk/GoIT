```python
df_2 = pd.read_csv('migr_asyappctza_page_tabular.tsv', sep = '\t')
df_2.head()
```

```python
df_3 = pd.concat([df_1["Country"], df_2.iloc[:, 1:]], axis=1)
df_3
```

```python
df_3.columns
```

```python
df_4 = df_3[df_3['Country'].isin(['EU27_2020','DE','EL','FR','IT'])]
df_4
```

```python
df_4.info()
```

```python
df_4.iloc[:, 1:] = df_4.iloc[:, 1:].astype(int)
df_4.info()
```

```python
df_4['2014 '] = df_4['2014 '].astype(int)
df_4['2015 '] = df_4['2015 '].astype(int)
df_4['2016 '] = df_4['2016 '].astype(int)
df_4['2017 '] = df_4['2017 '].astype(int)
df_4['2018 '] = df_4['2018 '].astype(int)
df_4['2019 '] = df_4['2019 '].astype(int)
df_4['2020 '] = df_4['2020 '].astype(int)
df_4['2021 '] = df_4['2021 '].astype(int)
df_4['2022 '] = df_4['2022 '].astype(int)
df_4['2023 '] = df_4['2023 '].astype(int)
print(df_4.dtypes)
```

```python
df_4.loc[[6:18], '2014 ':'2023 '] = df_4.loc[6:18, '2014 ':'2023 '].astype(int)
print(df_4.dtypes)
```

```python
df_4.index
```

```python
df_4.loc[df_4.index, df_4.columns[1:]] = df_4.loc[df_4.index, df_4.columns[1:]].astype(int)
print(df_4.dtypes)
```

```python
import pandas as pd
```

```python
asylum = pd.read_excel("Total Asylum.xlsx")
asylum
```

```python
asylum.info()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))
plt.plot(asylum['Year'], asylum['European Union'])
plt.plot(asylum['Year'], asylum[['Germany', 'Greece', 'France', 'Italy']], linestyle='dashed')
plt.legend(['European Union', 'Germany', 'Greece', 'France', 'Italy'], loc='upper center')
plt.title('Total asylum applicants', fontsize='large', fontweight='bold', loc='center')
plt.xlabel('Year', fontsize='medium', fontstyle='italic')
plt.ylabel('Millions of people', fontsize='medium', fontstyle='italic')
plt.show()
```

```python
asylum['Other'] = asylum['European Union'] - asylum['Germany'] - asylum['Greece'] - asylum['Italy'] - asylum['France']
asylum
```

```python
asylum_2023 = asylum.iloc[9,:]
asylum_2023
```

```python
print(type(asylum_2023))
```

```python
asylum_2023_1 = asylum_2023.reset_index()
asylum_2023_1
```

```python
asylum_2023_2 = asylum_2023_1.iloc[2:7, :]
asylum_2023_2
```

```python
asylum_2023_2.columns = ['Country', 'Person']
asylum_2023_2
```

```python
plt.figure(figsize=(7, 7))
plt.pie(
    x=asylum_2023_2['Person'],
    labels=asylum_2023_2['Country'],
    explode=[0, 0, 0, 0, 0.15],
    autopct="%.2f%%",
    pctdistance=0.75,
    labeldistance=1.15)

# Заголовок
plt.title('Share of Asylum applicants in 2023')
plt.show()
```

```python
import pandas as pd
```

```python
df_1 = pd.read_csv('migr_asyappctza_page_tabular.tsv')
df_1.head()
```

```python
df_1.iloc[0, -1].split("\t")[0]
```

```python
for index, row in df_1.iterrows():
    df_1.iloc[index, -1].split("\t")[0]
    print(df_1.iloc[index, -1].split("\t")[0])
```

```python
df_1["Country"] = "Country"
df_1.head()
```

```python
for index, row in df_1.iterrows():
    country = df_1.iloc[index, 6].split("\t")[0]
    df_1.iloc[index, 7] = country
df_1.head()
```

```python
df_2 = pd.read_csv('migr_asyappctza_page_tabular.tsv', sep = '\t')
df_2.head()
```

```python
df_3 = pd.concat([df_1["Country"], df_2.iloc[:, 1:]], axis=1)
df_3
```

```python
df_3.columns
```

```python
df_4 = df_3[df_3['Country'].isin(['EU27_2020','DE','EL','FR','IT'])]
df_4
```

```python
df_4.info()
```

```python
df_4.iloc[:, 1:] = df_4.iloc[:, 1:].astype(int)
df_4.info()
```

```python
df_4['2014 '] = df_4['2014 '].astype(int)
df_4['2015 '] = df_4['2015 '].astype(int)
df_4['2016 '] = df_4['2016 '].astype(int)
df_4['2017 '] = df_4['2017 '].astype(int)
df_4['2018 '] = df_4['2018 '].astype(int)
df_4['2019 '] = df_4['2019 '].astype(int)
df_4['2020 '] = df_4['2020 '].astype(int)
df_4['2021 '] = df_4['2021 '].astype(int)
df_4['2022 '] = df_4['2022 '].astype(int)
df_4['2023 '] = df_4['2023 '].astype(int)
print(df_4.dtypes)
```

```python
df_4.loc[[6:18], '2014 ':'2023 '] = df_4.loc[6:18, '2014 ':'2023 '].astype(int)
print(df_4.dtypes)
```

```python
df_4.index
```

```python
df_4.loc[df_4.index, df_4.columns[1:]] = df_4.loc[df_4.index, df_4.columns[1:]].astype(int)
print(df_4.dtypes)
```

```python
df_5 = df_4.sort_values(by='2023 ')
df_5
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
asylum = pd.read_excel("Total Asylum.xlsx")
asylum
```

```python
asylum.info()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(asylum['Year'], asylum['European Union'])
plt.plot(asylum['Year'], asylum[['Germany', 'Greece', 'France', 'Italy']], linestyle='dashed')
plt.legend(['European Union', 'Germany', 'Greece', 'France', 'Italy'], loc='upper center')
plt.title('Total asylum applicants', fontsize='large', fontweight='bold', loc='center')
plt.xlabel('Year', fontsize='medium', fontstyle='italic')
plt.ylabel('Millions of people', fontsize='medium', fontstyle='italic')
plt.show()
```

```python
asylum['Other'] = asylum['European Union'] - asylum['Germany'] - asylum['Greece'] - asylum['Italy'] - asylum['France']
asylum
```

```python
asylum_2023 = asylum.iloc[9,:]
asylum_2023
```

```python
asylum_2023_1 = asylum_2023.reset_index()
asylum_2023_1
```

```python
asylum_2023_2 = asylum_2023_1.iloc[2:7, :]
asylum_2023_2
```

```python
asylum_2023_2.columns = ['Country', 'Person']
asylum_2023_2
```

```python
plt.figure(figsize=(7, 7))
plt.pie(
    x=asylum_2023_2['Person'],
    labels=asylum_2023_2['Country'],
    explode=[0, 0, 0, 0, 0.15],
    autopct="%.2f%%",
    pctdistance=0.75,
    labeldistance=1.1)

# 
plt.title('Share of Asylum applicants in 2023')
plt.show()
```

```python
main_table = pd.read_excel("Main_table.xlsx", header=1)
main_table
```

```python
name_list = main_table.columns.to_list()
for name in name_list:
    print(name)
```

```python
employed_list = [name for name in name_list if name.startswith('Employed ICT specialists')]
employed_list
```

```python
main_table[['Country', 'Year'] + employed_list]
```

```python
digital_list = [name for name in name_list if name.startswith('Employed ICT specialists') or name.startswith('Individuals with basic digital skills')]
digital_list
```

```python
main_table[['Country', 'Year'] + digital_list]
```

```python
main_table[main_table['Year'] == 2021] [['Country', 'Year'] + digital_list]
```

```python
ict_df = main_table[main_table['Year'] == 2021] [['Country', 'Year'] + digital_list].copy()
```

```python
ict_corr_df = ict_df.iloc[:, 4:].corr().round(2)
ict_corr_df
```

```python
sns.heatmap(ict_corr_df, annot=True, linewidth=0.5, cmap="coolwarm", vmin=-1, vmax=1)

plt.title('Correlation heat map', fontsize='large', fontweight='bold', loc='center')
plt.show()
```

```python
education_list = [name for name in name_list if name.startswith('Tertiary education') or name.startswith('Employment rates')]
education_list
```

```python
edu_df = main_table[main_table['Year'].isin([2017, 2018, 2019, 2020, 2021, 2022, 2023])] [['Country', 'Year'] + education_list].copy()
edu_df
```

```python
edu_corr_df = edu_df.loc[edu_df['Country']=='France', education_list].corr().round(2)
edu_corr_df
```

```python
sns.heatmap(edu_corr_df, annot=True, linewidth=0.5, cmap="coolwarm", vmin=-1, vmax=1)

plt.title('Correlation heat map', fontsize='large', fontweight='bold', loc='center')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(edu_df[edu_df['Country']=='France']['Year'], edu_df[edu_df['Country']=='France']['Employment rates (from 25 to 49 years) \nMales (born in non-EU country), %'])
plt.plot(edu_df[edu_df['Country']=='France']['Year'], edu_df[edu_df['Country']=='France']['Employment rates (from 25 to 49 years) \nFemales (born in non-EU country), %'])
plt.plot(edu_df[edu_df['Country']=='France']['Year'], edu_df[edu_df['Country']=='France']['Tertiary education (from 25 to 34 years) \nMales (born in non-EU country), %'])
plt.plot(edu_df[edu_df['Country']=='France']['Year'], edu_df[edu_df['Country']=='France']['Tertiary education (from 25 to 34 years) \nFemales (born in non-EU country), %'])
plt.legend(['Employment rate M non-EU', 'Employment rate F non-EU', 'Tertiary education M non-EU', 'Tertiary education F non-EU'])
plt.show()
```

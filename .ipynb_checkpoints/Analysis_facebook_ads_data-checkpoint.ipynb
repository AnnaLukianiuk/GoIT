```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fb_data = pd.read_csv('facebook_ads_data.csv', parse_dates=[0])
display(fb_data.head())
fb_data.info()
```

```python
# 1.1 Зроби групування даних по дням та для згрупованих даних побудуй графік з щоденною сумою витрат на рекламу в 2021 році

fb_group = fb_data.groupby('ad_date', as_index=False).agg({
        'total_spend': 'sum',
        'total_impressions': 'sum',
        'total_clicks': 'sum',
        'total_value': 'sum'})

fb_group_21 = fb_group[(fb_group['ad_date'] >= '2021-01-01') & (fb_group['ad_date'] <= '2021-12-31')]
display(fb_group_21)
```

```python
plt.figure(figsize=(10, 5))

plt.plot(fb_group_21['ad_date'], fb_group_21['total_spend'])

plt.title('Total spend by date', fontsize='large', fontweight='bold', loc='center')
plt.xlabel('Date', fontsize='medium', fontweight='bold')
plt.ylabel('Total spend, USD', fontsize='medium', fontweight='bold')
plt.legend(['total_spend'], loc='upper right')
plt.show()
```

```python
fb_group_21.insert(column='rolling_total_spend', loc=2, value=fb_group_21['total_spend'].rolling(window=5).mean())

plt.figure(figsize=(10, 5))

plt.plot(fb_group_21['ad_date'], fb_group_21['rolling_total_spend'])

plt.title('Rolling total spend by date', fontsize='large', fontweight='bold', loc='center')
plt.xlabel('Date', fontsize='medium', fontweight='bold')
plt.ylabel('Rolling total spend, USD', fontsize='medium', fontweight='bold')
plt.legend(['rolling_total_spend'], loc='upper right')
plt.show()
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(15, 4))

ax[0].plot(fb_group_21['ad_date'], fb_group_21['total_spend'])
ax[0].set_title('Total spend by date')

ax[1].plot(fb_group_21['ad_date'], fb_group_21['rolling_total_spend'])
ax[1].set_title('Rolling total spend by date')

plt.show()
```

```python
# 1.2 Зроби групування даних по дням та для згрупованих даних побудуй графік з щоденним ROMI в 2021 році

plt.figure(figsize=(10, 5))

fb_group_21.insert(column='new_romi', loc=6, value=(fb_group_21['total_value']/fb_group_21['total_spend']))
plt.plot(fb_group_21['ad_date'], fb_group_21['new_romi'], 'r-', label='ROMI')

fb_group_21.insert(column='rolling_romi', loc=7, value=fb_group_21['new_romi'].rolling(window=7).mean())
plt.plot(fb_group_21['ad_date'], fb_group_21['rolling_romi'], 'b-', label='Rolling_ROMI')

plt.legend()
plt.xlabel('Date', fontsize='medium', fontweight='bold')
plt.ylabel('ROMI', fontsize='medium', fontweight='bold')
plt.title('ROMI by date', fontsize='large', fontweight='bold', loc='center')

plt.show()
```

```python
# 2.1 Зроби групування даних за назвою кампанії та побудуй графік загальної суми витрат на рекламу в кожній з кампаній

plt.figure(figsize=(15, 5))

sns.barplot(data=fb_data, x='campaign_name', y='total_spend', estimator='sum')

plt.title('Total spend by campaign', fontsize='large', fontweight='bold', loc='center')
plt.xlabel('Campaign name', fontstyle='italic', fontweight='bold')
plt.ylabel('Total spend, USD', fontstyle='italic', fontweight='bold')
plt.show()
```

```python
spend_by_campaign = fb_data.groupby('campaign_name', as_index=False)['total_spend'].sum().sort_values(by='total_spend')

plt.figure(figsize=(15, 5))

plt.bar(spend_by_campaign['campaign_name'], 
        spend_by_campaign['total_spend'],
        width=0.8)

plt.title('Total spend by campaign', fontsize='large', fontweight='bold', loc='center')
plt.xlabel('Campaign name', fontstyle='italic', fontweight='bold')
plt.ylabel('Total spend, USD', fontstyle='italic', fontweight='bold')

plt.ylim(0, spend_by_campaign['total_spend'].max()*1.1)

spend_by_campaign = spend_by_campaign.set_index('campaign_name')
for campaign in spend_by_campaign.index:
    plt.text(campaign, 
             spend_by_campaign.loc[campaign, 'total_spend'], 
             '$'+str(round(spend_by_campaign.loc[campaign, 'total_spend'], )), 
             ha='center',
             va='bottom')
plt.show()
```

```python
spend_by_campaign_daily = fb_data.groupby(['ad_date', 'campaign_name'], as_index=False)['total_spend'].sum()

plt.figure(figsize=(15, 7))
sns.lineplot(spend_by_campaign_daily, x='ad_date', y='total_spend', hue='campaign_name')
plt.title('Daily total spend by campaign', fontsize='large', fontweight='bold', loc='center')
plt.xlabel('Date', fontsize='medium', fontweight='bold')
plt.ylabel('Total spend, USD', fontsize='medium', fontweight='bold')
plt.show()
```

```python
# 2.2 Зроби групування даних за назвою кампанії та побудуй графік "Загальний ROMI в кожній з кампаній"

plt.figure(figsize=(15, 5))

sns.barplot(fb_data, x='campaign_name', y='romi')

plt.title('Average ROMI by campaign', fontsize='large', fontweight='bold', loc='center')
plt.xlabel('Campaign name', fontstyle='italic', fontweight='bold')
plt.ylabel('ROMI', fontstyle='italic', fontweight='bold')
plt.show()
```

```python
romi_by_campaign = fb_data.groupby('campaign_name', as_index=False)['romi'].mean().sort_values(by='romi')

plt.figure(figsize=(15, 5))

plt.bar(romi_by_campaign['campaign_name'], 
        romi_by_campaign['romi'],
        width=0.7)

plt.title('Average ROMI by campaign', fontsize='large', fontweight='bold', loc='center')
plt.xlabel('Campaign name', fontstyle='italic', fontweight='bold')
plt.ylabel('ROMI', fontstyle='italic', fontweight='bold')

plt.ylim(0, romi_by_campaign['romi'].max()*1.1)

romi_by_campaign = romi_by_campaign.set_index('campaign_name')
for romi in romi_by_campaign.index:
    plt.text(romi, 
             romi_by_campaign.loc[romi, 'romi'], 
             str(round(romi_by_campaign.loc[romi, 'romi'], 2)), 
             ha='center',
             va='bottom')
plt.show()
```

```python
# 3. За допомогою box plot визнач, яким був розкид щоденного ROMI в кожній з кампаній

plt.figure(figsize=(8, 6))

sns.boxplot(fb_data, x='romi', y='campaign_name')

plt.title('ROMI by campaign', fontsize='large', fontweight='bold', loc='center')
plt.xlabel('ROMI', fontsize='medium', fontweight='bold')
plt.ylabel('Campaign name', fontsize='medium', fontweight='bold')
plt.show()
```

```python
# 4. Побудуй гістограму з розподілом значень ROMI

sns.histplot(fb_data, x='romi')

plt.title('Histogram with the distribution of ROMI values', fontsize='large', fontweight='bold', loc='center')
plt.xlabel('ROMI', fontsize='medium', fontweight='bold')
plt.ylabel('Number of values', fontsize='medium', fontweight='bold')
plt.show()
```

```python
# 5. Побудуй теплову карту кореляції між усіма числовими показниками. 
# Які показники мають найвищу та найнижчу кореляцію? З чим корелює "total_value"?

fb_data_corr = fb_data.loc[:, 'total_spend':].corr().round(2)
display(fb_data_corr)
```

```python
plt.figure(figsize=(8, 5))

sns.heatmap(fb_data_corr, annot=True, linewidth=0.5, cmap="crest_r")

plt.title('Correlation heat map', fontsize='large', fontweight='bold', loc='center')
plt.show()
```

Найвищу кореляцію мають total value і total spend: 0.98

Найнижчу кореляцію мають ROMI і total value: -0.01

Total value корелює з: total spend (сильний зв'язок), total impressions і total clicks (помітний зв'язок)

```python
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(fb_data.corr(numeric_only=True), annot=True, fmt='.0%', cmap= 'crest_r')
```

```python
# 6. Побудуй точкову діаграму з лінійною регресією на даних з "total_spend" та "total_value"

sns.lmplot(fb_data, x='total_spend', y='total_value', scatter_kws={'color':'g'})

plt.title('Scatter plot with linear regression', fontsize='large', fontweight='bold', loc='center')
plt.xlabel('Total spend, USD', fontsize='medium', fontweight='bold')
plt.ylabel('Total value, USD', fontsize='medium', fontweight='bold')
plt.show()
```
